### sum / mean / axis /  keepdims

```python
# np.sum / ndarray.sum(axis=x)
import numpy as np

a = np.arange(6).reshape((2, -1))
print('ndarray : \n', a)
print('np.sum : ', np.sum(a))
print('ndaray.sum : ', a.sum())

print('======================2D==========================')
# axis argument
a = np.arange(12).reshape((3, -1))
sum_ = a.sum(axis = 0)
sum_1 = a.sum(axis = 1)
# n_dimension : (axis = 0, axis = 1, axis = . . ., axis = n-1)

print('ndarray : {}\n{}'.format(a.shape, a))                              
print('ndarray.sum(axis = 0) : {}\n{}'.format(sum_.shape, sum_))  # axis 와 남는 vector shape 생각하기 
print('ndarray.sum(axis = 1) : {}\n{}'.format(sum_1.shape, sum_1)) # axis로 설정한 차원이 사라진다고 생각?

                ########## (3, )은 바로 broadcasting이 안되니까 (3, 1)로 바로 만들자 ==>> keepdims ##########################
    
print('========================keepdims=============================')

a = np.arange(12).reshape((3, -1))
sum_class = np.sum(a, axis = 0, keepdims = True)
sum_student = np.sum(a, axis = 1, keepdims = True)

print("scores : {}\n{}".format(a.shape, a))
print("sum_class : {}\n{}".format(sum_class.shape, sum_class))
print('sum_student : {}\n{}'.format(sum_student.shape, sum_student))
```

```python
# mean substraction
import numpy as np

n_student, n_class = 3, 4 
m_score, M_score = 0, 100
scores = np.random.randint(low = m_score,
                          high = M_score,
                          size = (n_student, n_class))

mean_class = np.mean(scores, axis = 0, keepdims = True)
mean_student = np.mean(scores, axis = 1, keepdims = True)

print('scores : \n', scores)

print('shapes : \n', scores.shape, mean_class.shape, mean_student.shape, '\n')
print('Mean substraction by classes\n', scores - mean_class)
print('Mean substraction by student\n', scores - mean_student)
```

### 3rd order tensor case

```python
import numpy as np 

print('=======================axis = 0=====================')
a = np.arange(2 * 3 * 4).reshape((2, 3, 4))

sum_ = a.sum(axis = 0, keepdims = True)
print('a : \n', a)
print('\nsum_ (axis=1): \n', sum_)
print('\nsum_shape : ', sum_.shape)
mean_ = np.mean(a, axis = 0, keepdims = True)
print('\nmean : \n', mean_)
print('\nmean_shape :', mean_.shape)

print('=========================axis = 1====================')
a = np.arange(2 * 3* 4).reshape((2, 3, 4))
sum_ = a.sum(axis = 1, keepdims = True) ## T/F 바꿔가면서 차원 차이 확인하기 
print('sum_ (axis=1):\n', sum_)
print('\nsum_shape : ',sum_.shape) # (2, 1, 4)

print('========================axis = 2=====================')
sum_ = a.sum(axis = 2, keepdims = True)
print('sum_ (axis=2):\n', sum_)
print('\nsum_shape : ', sum_.shape)
```

### sum / cumsum

```python
# numpy.cumsum / ndarray.cumsum(a, axis = None , dtype = None ,out = None )  : 누적 합 
# sum은 scalar 값을 반환하지만, cumsum은 vector의 크기를 유지한다
import numpy as np 

a = np.arange(5)
print("ndarray : ", a)
cumsum = np.cumsum(a)
print('cumsum : ', cumsum)
# bar graph로 시각화하기 좋다

print('========================2D====================')

a = np.arange(3 *3).reshape((3, 3))
print("ndarray : {}\n{}".format(a.shape, a))

cumsum = np.cumsum(a)
print("\ncumsum : {}\n{}".format(cumsum.shape, cumsum)) # flatten 뒤에 cumsum한 것과 같다

cumsum = np.cumsum(a).reshape((3, 3))
print("\ncumsum : {}\n{}".format(cumsum.shape, cumsum))

print('==============================================')

import numpy as np

a = np.arange(3*4).reshape((3, -1))
print("ndarray : {}\n{}\n".format(a.shape, a))

cumsum = np.cumsum(a, axis = 0)
print("\ncumsum (axis = 0): {}\n{}".format(cumsum.shape, cumsum))
print("\ncumsum[::-1, :] (axis = 0): {}\n{}".format(cumsum.shape, cumsum[::-1, :]))
cumsum = np.cumsum(a, axis = 1)
print("\ncumsum (axis = 1): {}\n{}".format(cumsum.shape, cumsum))
print
```

### prod / cumprod

```python
# np.prod, np.cumprod 

import numpy as np

a = np.arange(1,5)
print("ndarray : {}\n{}\n".format(a.shape, a))

prod = np.prod(a)
cumprod = np.cumprod(a)

print("prod : {}\n{}\n".format(prod.shape, prod))
print("cumprod : {}\n{}\n".format(cumprod.shape, cumprod))

print("===================2D======================")

a = np.arange(1, 13).reshape((3, 4))
print("ndarray : {}\n{}\n".format(a.shape, a))

prod = np.prod(a, axis = 0)
cumprod = np.cumprod(a, axis = 0)

print("prod : {}\n{}\n".format(prod.shape, prod))
print("cumprod : {}\n{}\n".format(cumprod.shape, cumprod))
```

```python
# var, std
import numpy as np 

scores = np.random.normal(loc = 10, scale=5, size = (100, ))
var = scores.var()
std = scores.std()
var= np.var(scores)
std = np.std(scores)

print(var, std)

print('======================================')
# standardization 
means = [50, 60, 70]
stds = [3, 5, 10]
n_student, n_class = 100, 3

scores = np.random.normal(loc=means,scale=stds, size=(n_student, n_class))
scores = scores.astype(np.float32)

print("shape of scores : ", scores.shape)
print("dtype of scores : ", scores.dtype)

means = scores.mean(axis = 0)
stds = scores.std(axis= 0)

print("means before stdz : \n", means)
print("stds before stdz : \n", stds)

score_stdz = (scores - means) / stds
means_stdz = score_stdz.mean(axis = 0)
stds_stdz = score_stdz.std(axis = 0)
print("scores after stdz : \n",score_stdz)
print("means after stdz : \n",means_stdz)
print("stdz after stdz : \n",stds_stdz)
```

### Max, Min

```python
# np.amax / ndarray.amax / numpy.argmax / ndarray.argmax

import numpy as np
a = np.random.randint(0, 100, (10, ))

M = np.max(a)
M_idx = np.argmax(a) # Max 값의 index

print(f"ndarray : \n{a}")
print(f"M / M_idx : {M}/{M_idx}")

means = [50, 60, 70]
stds = [3, 5, 10]
n_student, n_class = 100, 3 

scores = np.random.normal(loc = means, scale = stds, size = (n_student, n_class))
scores = scores.astype(np.float32)

scores_max = np.amax(scores, axis = 0, keepdims = True)
scores_max_idx = np.argmax(scores, axis = 0)
print("max scores : ", scores_max, scores_max.shape)
print("Max indices : ", scores_max_idx)

# 학생별로 시험 잘 본 과목 
score_max_idx = np.argmax(scores, axis = 1)
print("Max scores subject : ",score_max_idx)
```

```python
# numpy.amin() / ndarray.amin
# Min - Max normalization

import numpy as np
means = [50, 60, 70]
stds = [3, 5, 10]
n_student, n_class = 100, 3

scores = np.random.normal(loc=means,scale=stds,size=(n_student,n_class))
scores = scores.astype(np.float32)

scores_max = np.amax(scores, axis = 0)
scores_min = np.amin(scores, axis = 0)

scores_mM_norm = (scores - scores_min) / (scores_max - scores_min)
print(scores_mM_norm)
```

### Rounding

```python
# numpy.around / numpy.round_ / ndarray.round(decimals = 0, out = None)
# decimals : 표기할 자리수 --- decimals = 2 일때, 셋째자리에서 반올림 
# decimals 는 기본 0이므로 첫째자리에서 반올림한 정수부분만 표현 

import numpy as np

x = np.random.uniform(-5, 5, (5, ))

np_around = np.around(x, decimals=2)
x_round = x.round(decimals = 2)  

print(x)
print(np_around)
print(x_round)

print("\n")

scores = np.random.uniform(0, 100, (100, 5))
means = scores.mean(axis=0)
stds = scores.std(axis =0)

print(f'class means : {means.round(2)}\n')
print(f'class stds : {stds.round(2)}')
```

```python
# numpy.ceil / numpy.floor : 소숫점 첫째자리에서 올림, 내림

import numpy as np

x = np.random.uniform(-5, 5, (5, ))

ceil = np.ceil(x)
floor = np.floor(x)

print(f'x : \n{x}\n')
print(f'ceil : \n{ceil}\n')
print(f'floor : \n{floor}\n')

print("===================")

# numpy.trunc : 소숫점 첫째자리까지 그냥 날리고 정수부분만 
x = np.random.uniform(-5, 5, (5, ))

trunc = np.trunc(x)
print(f'x : \n{x}\n')
print(f'trunc : \n{trunc}')

trunc_where = np.where(x >=0, np.floor(x), np.ceil(x))
print(f'trunc_where : \n{trunc_where}\n')

# 둘째자리에서 버리기

trunc = 0.1 * np.trunc(10*x)
print(f'trunc(0.1) : \n{trunc}\n')

# 정수부, 소수부 나누기
int_part = np.trunc(x)
frac_part = x - int_part
print(f"frac_part = \n{frac_part}\n")
```

### Sorting

```python
# numpy.sort() / numpy.argsort()
import numpy as np

x = np.random.randint(0, 20, (20, ))

sorted_x = np.sort(x)
argsort_x = np.argsort(x)

print(f'origin : {x}\n')
print(f'sort : {sorted_x}\n')
print(f'argsort_x : {argsort_x}\n')

sort = np.sort(x)[::-1]
argsort = np.argsort(x)[::-1]
print(f'sort : {sort}')

# 
print('================================')
pred = np.random.uniform(0, 100, (5, ))
pred /= pred.sum()

top3_pred = np.sort(pred)[::-1][:3]
top3_indices = np.argsort(pred)[::-1][:3]

print(f'pred : \n{pred.round(3)}\n')
print(f'top3_pred : \n{top3_pred.round(3)}')
print(f'top-3 indices : \n{top3_indices}')
```

```python
# sorting matrix  (row)
import numpy as np

x = np.random.randint(0, 100, (4, 5))

sort = np.sort(x, axis = 0)
argsort = np.argsort(x, axis = 0)

print(f'x : \n{x}\n')
print(f'sort_x : \n{sort}\n')
print(f'argsort : \n{argsort}\n')

sort = sort[:, 0]
print(f'sort_x : \n{sort}\n')  # 첫번째 행만 뽑기

print('====================================================')

sort = np.sort(x, axis = 0)[::-1, :]
argsort = np.argsort(x, axis = 0)[::-1, :]

print(f'x : \n{x}\n')
print(f'sort_x : \n{sort}\n')
print(f'argsort_x : \n{argsort}\n')
```

```python
# sorting matirx

import numpy as np 

scores = np.random.randint(0, 100, (5, 3))
sort = np.sort(scores, axis = 0)[::-1, :]
argsort = np.argsort(scores, axis = 0)[::-1, :]
top2_scores = sort[:2, :]
top2_students = argsort[:2, :]

print(f'scores : \n{scores}\n')
print(f'sort : \n{sort}\n')
print(f'argsort : \n{argsort}\n')
print(f'top2-scores : \n{top2_scores}\n')
print(f'top2-students : \n{top2_students}\n')

print("=========================================")

# column 

x = np.random.randint(0, 100, (4, 5))

sort_ascending = np.sort(x, axis = 1)
argsort_ascending = np.argsort(x, axis = 1)

sort_decending = np.sort(x, axis = 1)[:, ::-1]
argsort_decending = np.argsort(x, axis = 1)[:, ::-1]

print(f'x : \n{x}\n')
print(f'ascending : \n{sort_ascending}\n')
print(f'argsort_ascending : \n{argsort_ascending}\n')
print(f'decending : \n{sort_decending}\n')
print(f'argsort_decending : \n{argsort_decending}\n')
```
